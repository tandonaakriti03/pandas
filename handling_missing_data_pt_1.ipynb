```python
import pandas as pd
df=pd.read_csv("/content/weather_data (1).csv",parse_dates=["day"]) #change the data type for day
type(df.day[0]) #check the data type
```

```python
df.set_index('day',inplace=True)#set the day as index
df
```

```python
new_df=df.fillna(0)
new_df
```

```python
new_df=df.fillna({
    'temperature': 0,
    'windspeed':0,
    'event':'no event'
})
new_df
```

```python
new_df=df.fillna(method='ffill') # carry forward previous value
new_df
```

```python
new_df=df.fillna(method='ffill', limit=1) # limit is used to copy previous value for specific record only
new_df
```

```python
new_df=df.fillna(method='bfill') #fill value forward value
new_df
```

```python
new_df=df.fillna(method='bfill',axis='columns') #fill value column wise
new_df
```

```python
new_df=df.interpolate() #a method for generating points between given points.
new_df
```

```python
new_df=df.interpolate(method='time') #a method for generating points between given points.
new_df
```

```python
new_df=df.dropna()#to drop all na value even if row has one na value
new_df
```

```python
new_df=df.dropna(how='all')#it drop the row if has ass the value as na
new_df
```

```python
new_df=df.dropna(thresh=1)#if you have atleast 1 non na value then keep that row and drop any other row
new_df
```

```python
#how to add missing dates
dt=pd.date_range("01-01-2017","01-11-2017") #date range
idx=pd.DatetimeIndex(dt) #creating date time index
df=df.reindex(idx) # re index data set
df
```
