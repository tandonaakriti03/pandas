```python
import pandas as  pd
import numpy as np
df=pd.read_csv("/content/weather_data.csv")
df
```

```python
new_df=df.replace(-99999,np.NaN) # to replace the value and for more than signle value  to be replaced  (-9999,-8888)
new_df
```

```python
new_df=df.replace({
    'temperature':-99999,
    'windspeed':-99999,
    'event':'0'
},np.NaN)
new_df
```

```python
new_df=df.replace({
    -99999:np.NaN,
    '0' : 'sunny'
})
new_df
```

```python
 new_df=df.replace({
     'temperature': '[A-Za-z]',
     'windspeed': '[A-Za-z]',
 },'',regex=True)
 new_df
```
