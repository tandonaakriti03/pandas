```python
import  pandas as pd
df1=pd.DataFrame({
    "city":["new york","chicago","orlando"],
    "temperature":[21,14,35],
})
df1
```

```python
df2=pd.DataFrame({
    "city":["new york","chicago","orlando"],
    "humdity":[65,68,75],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city") # removes index usage
df3
```

If a datasets have uncommon entries then we use HOW="OUTER"

```python
df1=pd.DataFrame({
    "city":["new york","chicago","orlando"," baltimore"],
    "temperature":[21,14,35,34],
})
df1
```

```python
df2=pd.DataFrame({
    "city":["new york","chicago","san franciso"],
    "humdity":[65,68,75],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city",how="outer",indicator=True)
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="inner")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="left")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="right")
df3
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3=pd.merge(df1,df2,on="city",suffixes=('_left','_right'))
df3
```
