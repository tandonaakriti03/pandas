```python
import pandas as pd
india_weather=pd.DataFrame({
    "city":["mumbai","delhi","banglore"],
    "temperatur":[32,45,30],
    "humidity":[80,60,90]
})
india_weather
```

```python
us_weather=pd.DataFrame({
    "city":["new york ","chicago","orlando"],
    "temperatur":[21,14,35],
    "humidity":[68,65,75]
})
us_weather
```

```python
df=pd.concat([india_weather,us_weather])
df
```

```python
df=pd.concat([india_weather,us_weather],ignore_index=True) #helps to get index in format
df
```

```python
df=pd.concat([india_weather,us_weather],keys=["india","us"]) #helps to get index in format
df
```

```python
temperature_df=pd.DataFrame({
    "City": ["mumbai","delhi","banglore"],
    "temperature":[32,45,30],
},index=[0,1,2])
temperature_df
```

```python
df=pd.concat([temperature_df,windspeed_df])
df
```

```python
df=pd.concat([temperature_df,windspeed_df], axis=1)
df
```
